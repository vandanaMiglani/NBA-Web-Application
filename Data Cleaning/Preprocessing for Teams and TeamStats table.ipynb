```python
team_stats.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
team_stats.rename(columns={'WIN%':'Win_Percentage','FG%':'FG_Percentage','3P%':'3P_Percentage','FT%':'FT_Percentage','+/-':'Plus_Minus','PTS':'Pts','MIN':'Min','SEASON': 'Season'},inplace=True)
```

```python
team_stats.columns
```

```python
teams_df1 = teams_df[['Team_ID','Team_Name']]
teams_df1
```

```python
result = teams_df1.merge(team_stats,left_on="Team_Name",right_on="TEAM")
```

```python
result
```

```python
result.drop(columns=['TEAM','Team_Name'],inplace=True)
```

```python
result
```

```python
result['Season'] = result['Season'].apply(lambda x : x.split("-")[0])
```
